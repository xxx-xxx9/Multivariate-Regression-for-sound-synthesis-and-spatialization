```python
# 1500 punti 2D hanno lo stesso target. Ci sono 6 target e 30 volatili.

import json
import os
import numpy as np

def crea_json_da_file(path_txt, file_output_json):
    """
    Legge i dati da 6 file di testo, ciascuno con 30 righe di coppie (x,y),
    e crea un file JSON con 'mappings', 'labels' e 'readings'.

    Args:
        di_coppi_txt (str): percorso/cartella dove si trovano i file passo_1.txt, passo_2.txt ecc.
        file_output_json (str): nome o percorso del file JSON da scrivere.
    """

    # Dividiamo il piano cartesiano tra 0 e 1 in 2x3
    # selezioniamo per ciascuno dei 6 quadrati
    # 1500 punti a caso


    # Definiamo i parametri della griglia
    grid_size1 = 2  # 2x3 quadrati
    grid_size2 = 3  # 2x3 quadrati
    points_per_square = 250  # Numero di punti per ogni rettangolo
    
    # Lista per memorizzare le coordinate in ogni quadrato
    all_points = []
    square_points = []
    
    # Creazione dei punti casuali all'interno di ciascun rettangolo
    for ii in range(grid_size1):
        for jj in range(grid_size2):
            #square_points = []
            # Determiniamo i limiti del quadrato
            x_min, x_max = ii / grid_size1, (ii + 1) / grid_size1
            y_min, y_max = jj / grid_size2, (jj + 1) / grid_size2
            
            # Generiamo punti casuali all'interno di questi limiti
            x_coords = np.random.uniform(x_min, x_max, points_per_square)
            y_coords = np.random.uniform(y_min, y_max, points_per_square)
            
            # Creiamo la lista di coordinate (x, y) per questo quadrato
            #square_points = list(zip(x_coords, y_coords))
            for i in range(points_per_square):
                square_points.append([round(x_coords[i],2),round(y_coords[i],2)])
            
            # Aggiungiamo alla lista principale
            #all_points.append(square_points)
    print("lunghezza square_points", len(square_points))
    print("square_points", square_points)


    
    # Le 250 coppie di coordinate (x,y) che vengono associate ciascuna a 1 dei 6 file
    readings_list = square_points
    print("lunghezza readings_list", len(readings_list))
    print("readings_list ", readings_list)
    
    data = {
        "bidim": [],
       # "labels": [],
        "multidim": []
    }
    
    # Abbiamo 9 file: passo_1.txt, passo_2.txt, ..., passo_9.txt
    num_file = 6

    for i in range(1, num_file + 1):
        filename = os.path.join(path_txt, f"passo_{i}.txt")

        for ind in range(int(1500/num_file)): # ho lo stesso target/freeze per 1500/6=250 punti 2D
            coppie = []

            with open(filename, 'r') as f:
                for line in f:
                    line = line.strip()
                    if not line:
                        continue  # salta eventuali righe vuote
                    
                    # Splitta sulla virgola
                    x_str, y_str = line.split(',')
                    
                    # Converte in float
                    x_val = float(x_str.strip())
                    y_val = float(y_str.strip())
                    
                    # Aggiunge la coppia [x_val, y_val] alla lista
                    
                    coppie.append([x_val, y_val])
        
            # 'coppie' ora contiene 30 coppie (x,y) lette dal file
            data["multidim"].append(coppie)
        
        
        # Label: un identificatore numerico, ad esempio da 0 a 16
        #data["labels"].append(i - 1)
        
        # Inserisce la coppia di readings corrispondente
        # (i-1 perché l'indice parte da 0)
    data["bidim"] = readings_list
    print("lunghezza data bidim ", len(data["bidim"]))
    print("data bidim ", data["bidim"][0])
    print("multidim[150]", data["multidim"][150])
    
    # Salviamo il dizionario in un file JSON
    with open(file_output_json, 'w') as out:
        json.dump(data, out, indent=2)


if __name__ == "__main__":
    # cartella in cui si trovano i file testo (passo_1.txt, passo_2.txt, ...)
    folder_con_testi = "Data/punti_53D-1500_30x6agent_classOKok_seed12"
    # nome del file JSON di destinazione
    output_json = "MLP_Boyds1500_30agent_classOKok_seed12_6selFreeze.json"
    
    crea_json_da_file(folder_con_testi, output_json)
    print(f"File JSON creato: {output_json}")
```
